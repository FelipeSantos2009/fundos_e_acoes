```python
from selenium import webdriver

# passo 1: abrir o navegador
navegador = webdriver.Chrome()
navegador.get('https://www.google.com')
```

```python
# passo 2: importar a base de dados
import pandas as pd

tabela = pd.read_excel('commodities.xlsx')
display(tabela)
```

```python
#import unicodedata
#produto = unicodedata.normalize('NFKD', produto).encode('ascii', 'ignore')

for linha in tabela.index:
    produto = tabela.loc[linha, 'Produto']# loc para localizar o produto correspondente de cada linha da tabela
    print(produto)
    produto = produto.replace('ó', 'o').replace('ã', 'a').replace('á', 'a').replace('ç', 'c').replace('ú', 'u').replace('é', 'e')
    link = f'https://www.melhorcambio.com/{produto}-hoje'
    navegador.get(link)
    print(link)
# xpath é um código que todos   
#os elementos tem, para saber como 
#consegui-lo, você precisa inspecionar, clicar em cursor de mouse no canto superior esquerdo, clicar no elemento que 
#você quiser e copiar o código destacado em azul 
    preco = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    #.click
#send.keys('texto') => escrever
#get_attribute() -> pegar um valor
    preco.replace('.', '').replace(',', '.')
    print(preco)
    tabela.loc[linha, 'Preço Atual'] = preco # substitui os valores vazios com o preço dos produtos
tabela['Preço Atual'].replace(',', '.')
display(tabela)

#preencher a coluna de comprar
tabela['Comprar'] = tabela['Preço Atual'] < tabela['Preço Ideal']
display(tabela)
```

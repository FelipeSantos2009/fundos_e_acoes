```python
from selenium import webdriver
from time import sleep as sl
import pandas as pd
import pyautogui as py

df = pd.read_excel('projeto_bi_2.xlsx')
display(df)
```

```python
navegador = webdriver.Chrome()
navegador.get('https://www.google.com')
link = 'https://statusinvest.com.br/'
navegador.get(link)
print(link)

py.keyDown('win')
py.press('up')
py.keyUp('win')
cot_atual2 = []
preco_min2 = []
preco_max2 = []
div_yield2 = []
liq_diaria2 = []
p_vp2 = []
cagr2 = []
patr_liq2 = []
caixa2 = []
lucro2 = []
for c in df['AÇÕES']:
    #print(c)
    
    navegador.find_element('xpath', '//*[@id="main-nav-nav"]/div/div/div/ul/li[2]/a/i').click()
    navegador.find_element('xpath', '//*[@id="main-search"]/div[1]/span[1]/input[2]').send_keys(c)
    sl(3)
    navegador.find_element('xpath', '//*[@id="main-search"]/div[2]/div/div/a').click()
    sl(1.5)
    cot_atual = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div/div[1]/div/div[1]/div/div[1]/strong').text
    #print('COTAÇÃO ATUAL')
    #print(cot_atual)
    cot_atual2.append(cot_atual)
    #print()
    
    preco_min = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div/div[1]/div/div[2]/div/div[1]/strong').text
    #print('PREÇO MÍNIMO')
    #print(preco_min)
    preco_min2.append(preco_min)
    #print()
    
    preco_max = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div/div[1]/div/div[3]/div/div[1]/strong').text
    #print('PREÇO MÁXIMO')
    #print(preco_max)
    preco_max2.append(preco_max)
    #print()
    
    div_yield = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div/div[1]/div/div[4]/div/div[1]/strong').text
    #print('DIVIDEND YIELD')
    #print(div_yield)
    div_yield2.append(div_yield)
    #print()
    
    py.scroll(-1200)
    sl(0.8)
    liq_diaria = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div/div[5]/div/div/div[3]/div/div/div/strong').text
    # o .text pega o que não está dentro de uma variável "value"
    #print('liquidez diária')
    #print(liq_diaria)
    liq_diaria2.append(liq_diaria)
    #print()
    
    py.scroll(-700)
    p_vp = navegador.find_element('xpath', '//*[@id="indicators-section"]/div[2]/div/div[1]/div/div[4]/div/div/strong').text
    #print('p/vp')
    #print(p_vp)
    p_vp2.append(p_vp)
    #print()
    
    cagr = navegador.find_element('xpath', '//*[@id="indicators-section"]/div[2]/div/div[5]/div/div[2]/div/div/strong').text
    cagr2.append(cagr)
    
    navegador.find_element('xpath', '//*[@id="main-2"]/div[1]/div/div/ul/li[2]/a').click()
    sl(1)
    patr_liq = navegador.find_element('xpath', '//*[@id="company-section"]/div[1]/div/div[2]/div[1]/div/div/strong').text
    #print('PATRIMÔNIO LÍQUIDO')
    #print(patr_liq)
    patr_liq2.append(patr_liq)
    #print()
    
    navegador.find_element('xpath', '//*[@id="main-2"]/div[1]/div/div/ul/li[3]/a').click()
    sl(1)
    py.scroll(-1000)
    sl(2)
    caixa = navegador.find_element('xpath', '//*[@id="contabil-section"]/div[1]/div/div[3]/div[1]/div[1]/table/tbody/tr[1]/td[2]/span').text
    #print('CAIXA LÍQUIDO')
    #print(caixa)
    caixa2.append(caixa)
    #print()
    
    lucro = navegador.find_element('xpath', '//*[@id="contabil-section"]/div[1]/div/div[3]/div[1]/div[1]/table/tbody/tr[3]/td[2]/span').text
    #print('LUCRO LÍQUIDO')
    #print(lucro)
    lucro2.append(lucro)
    #print()
```

```python
#df = pd.DataFrame(df[''])
```

```python
df = pd.DataFrame(zip(cot_atual2, preco_min2, preco_max2, div_yield2, liq_diaria2, p_vp2, cagr2, patr_liq2, caixa2, lucro2), columns=['COTAÇÃO ATUAL', 'PREÇO MÍNIMO', 'PREÇO MÁXIMO', 'DIVIDEND YIELD', 'LIQUIDEZ DIÁRIA', 'P/VP', 'CAGR LUCRO', 'PATRIMÔNIO LÍQUIDO', 'CAIXA', 'LUCRO'])
display(df)
```

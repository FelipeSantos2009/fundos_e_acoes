```python
from selenium import webdriver
from time import sleep as sl
import pandas as pd
import pyautogui as py

df = pd.read_excel('Fundos1.xlsx')
display(df)
```

```python
navegador = webdriver.Chrome()
navegador.get('https://www.google.com')
link = 'https://statusinvest.com.br/'
navegador.get(link)
print(link)

py.keyDown('win')
py.press('up')
py.keyUp('win')
fundos2 = []
cot_atual2 = []
preco_min2 = []
preco_max2 = []
div_yield2 = []
p_vp = []
cagr = []
val_patr = []
caixa = []
liq_diaria = []

for c in df['FUNDOS']:
    navegador.find_element('xpath', '//*[@id="main-nav-nav"]/div/div/div/ul/li[2]/a/i').click()
    navegador.find_element('xpath', '//*[@id="main-search"]/div[1]/span[1]/input[2]').send_keys(c)
    sl(3.5)
    navegador.find_element('xpath', '//*[@id="main-search"]/div[2]/div/div/a').click()
    sl(1.8)
    
    cot_atual = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[1]/div[1]/div/div[1]/strong').text
    cot_atual2.append(cot_atual)
    
    preco_min = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[1]/div[2]/div/div[1]/strong').text
    preco_min2.append(preco_min)
    
    preco_max = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[1]/div[3]/div/div[1]/strong').text
    preco_max2.append(preco_max)
    
    div_yield = navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[1]/div[4]/div/div[1]/strong').text
    div_yield2.append(div_yield)
    
    fundos2.append(c)
    py.scroll(-1100)
    sl(0.8)
    try:
        p_vp.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[4]/div/div[2]/div/div[1]/strong').text)
    
    except:
        p_vp.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[5]/div/div[2]/div/div[1]/strong').text)
        

    try:
        cagr.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[4]/div/div[5]/div/div[1]/strong').text)
    
    except:
        cagr.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[5]/div/div[5]/div/div[1]/strong').text)
    
    try:
        val_patr.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[4]/div/div[1]/div/div[2]/span[2]').text)
    
    except:
        val_patr.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[5]/div/div[1]/div/div[2]/span[2]').text)
        
    try:
        caixa.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[4]/div/div[3]/div/div[2]/span[2]').text)
        
    except:
        caixa.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[5]/div/div[3]/div/div[2]/span[2]').text)
    
    
    py.scroll(-150)
    try:
        liq_diaria.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[5]/div/div/div[3]/div/div/div/strong').text)
        
    except:
        liq_diaria.append(navegador.find_element('xpath', '//*[@id="main-2"]/div[2]/div[6]/div/div/div[3]/div/div/div/strong').text)
    
    
```

```python
df = pd.DataFrame(zip(fundos2, cot_atual2, preco_min2, preco_max2, div_yield2, p_vp, cagr, val_patr, caixa, liq_diaria), columns=['FUNDOS', 'COTAÇÃO ATUAL', 'PREÇO MÍNIMO', 'PREÇO MÁXIMO', 'DIVIDEND YIELD', 'P_VP', 'CAGR LUCRO', 'PATRIMÔNIO', 'CAIXA', 'LIQUIDEZ DIÁRIA'])
display(df)
```

```python
df.to_excel('fundos_selenium_final8.xlsx')
```

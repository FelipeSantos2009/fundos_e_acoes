```python
from selenium import webdriver
from time import sleep
import pyautogui
import pyperclip

# ATIVIDADE 1: PEGAR COTAÇÃO DE 100 MIL MILHAS DE 3 COMPANHIAS PELA HOTMILHAS 

# email 1: latam pass
navegador = webdriver.Chrome()
navegador.get('https://www.google.com')
link = 'https://hotmilhas.com.br'
navegador.get(link)
print(link)

#cadastra o e-mail
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[1]/input').send_keys('overskytur@gmail.com')

#selecionar a compania aérea
sleep(3)
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[1]/span').click()

#latam pass
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[2]/div/ul/li[1]').click()

#selecionar quantidade de milhas
navegador.find_element('xpath', '//*[@id="select_points_2"]/div[1]').click()
sleep(1)

# 100 mil milhas
pyautogui.moveTo(x=868, y=572)
sleep(0.5)
pyautogui.scroll(-4700)
sleep(1)
navegador.find_element('xpath', '//*[@id="select_points_2"]/div[2]/div[2]/ul/li[89]').click()

#fechar operação
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[2]/div/div/button').click()
sleep(1)
pyautogui.click(x=982, y=34)
sleep(1)

#.click()
#send.keys('texto') => escrever
#get_attribute() -> pegar um valor
```

```python
# EMAIL 2: SMILES
#nova cotação
navegador.find_element('xpath', '//*[@id="formulario-cotacao"]/div/div/div/div/div[1]/div/a[1]').click()

#cadastra o e-mail
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[1]/input').send_keys('overskytur@gmail.com')

#selecionar a companhia aérea
sleep(2)
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[1]/span').click()

#smiles
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[2]/div/ul/li[3]').click()

#selecionar quantidade de milhas
navegador.find_element('xpath', '//*[@id="select_points_1"]/div[1]').click()
sleep(2)

# 100 mil milhas
pyautogui.moveTo(x=868, y=572)
sleep(0.5)
pyautogui.scroll(-2600)
sleep(1)
navegador.find_element('xpath', '//*[@id="select_points_1"]/div[2]/div[2]/ul/li[51]').click()

#fechar operação
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[2]/div/div/button').click()
pyautogui.click(x=982, y=34)
sleep(1)
```

```python
#EMAIL 3: TUDO AZUL
#nova cotação
navegador.find_element('xpath', '//*[@id="formulario-cotacao"]/div/div/div/div/div[1]/div/a[1]').click()

#cadastra o e-mail
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[1]/input').send_keys('overskytur@gmail.com')

#selecionar a companhia aérea
sleep(2)
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[1]/span').click()

#azul
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[1]/div[2]/div/div/div[2]/div/ul/li[4]').click()

#selecionar quantidade de milhas
navegador.find_element('xpath', '//*[@id="select_points_3"]/div[1]').click()
sleep(2)

# 100 mil milhas
pyautogui.moveTo(x=868, y=572)
sleep(0.5)
pyautogui.scroll(-4700)
sleep(1)
navegador.find_element('xpath', '//*[@id="select_points_3"]/div[2]/div[2]/ul/li[89]').click()

#fechar operação
navegador.find_element('xpath', '//*[@id="quotation-form"]/div[2]/div/div/button').click()
pyautogui.click(x=982, y=34)
sleep(1)
```
